```python
import pandas as pd
import os
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR

# The file was extracted to the 'dataset' folder
filepath = 'dataset/measures_v2.csv'

if 'df' not in locals():
    if os.path.exists(filepath):
        df = pd.read_csv(filepath)
        print(f"Loaded {filepath} successfully.")
    else:
        print(f"Error: '{filepath}' not found. Please ensure the file is extracted correctly.")

if 'df' in locals():
    # Define features and target
    X = df.drop(columns=['pm'])
    y = df['pm']

    # Split the data
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3)

    lr = LinearRegression()
    dr = DecisionTreeRegressor()
    rf = RandomForestRegressor()

    print("Starting training for Linear Regression...")
    lr.fit(X_train, y_train)
    print("Starting training for Decision Tree...")
    dr.fit(X_train, y_train)
    print("Starting training for Random Forest (this may take a while)... ")
    rf.fit(X_train, y_train)
    print("Training complete for available models.")
```

```python
from sklearn import metrics

# Generate predictions for each trained model
p1 = lr.predict(X_test)
p2 = dr.predict(X_test)
p3 = rf.predict(X_test)

print('Linear Regression R2:', metrics.r2_score(y_test, p1))
print('Decision Tree R2:', metrics.r2_score(y_test, p2))
print('Random Forest R2:', metrics.r2_score(y_test, p3))
```

```python
from sklearn.metrics import mean_squared_error

print(mean_squared_error(y_test, p1))
```

```python
import gradio as gr
import numpy as np

# Dummy prediction function (replace with your ML model later)
def predict_temp(ambient, coolant, vd, vq, speed, id_current, iq_current):
    # Simple demo formula
    prediction = (0.3*ambient + 0.2*coolant +
                  0.01*vd + 0.01*vq +
                  0.002*speed +
                  0.05*id_current + 0.05*iq_current)
    return round(prediction, 3)

with gr.Blocks(theme=gr.themes.Base()) as app:

    gr.Markdown(
        """
        <h1 style='text-align: center; color: black;'>
        Electric Motor Temperature Prediction
        </h1>
        <p style='text-align: center;'>
        Fill in the details below to predict Permanent Magnet Surface Temperature
        </p>
        """
    )

    with gr.Column():
        ambient = gr.Number(label="Ambient Temperature (C)")
        coolant = gr.Number(label="Coolant Temperature (C)")
        vd = gr.Number(label="Voltage d-component")
        vq = gr.Number(label="Voltage q-component")
        speed = gr.Number(label="Motor Speed (RPM)")
        id_current = gr.Number(label="Current d-component")
        iq_current = gr.Number(label="Current q-component")

    submit_btn = gr.Button("Predict", variant="primary")

    output = gr.Textbox(label="Permanent Magnet Surface Temperature (C)")

    submit_btn.click(
        predict_temp,
        inputs=[ambient, coolant, vd, vq, speed, id_current, iq_current],
        outputs=output
    )

app.launch(share=True)
```
